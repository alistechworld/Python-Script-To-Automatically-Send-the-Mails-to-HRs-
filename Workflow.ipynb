```python
import pandas as pd
import smtplib
from email.mime.multipart import MIMEMultipart
from email.mime.text import MIMEText
from email.mime.base import MIMEBase
from email import encoders
from google.colab import files

# === STEP 1: Upload Excel & CV ===
print("Upload Excel file:")
uploaded = files.upload()

print("Upload CV file:")
cv_upload = files.upload()

excel_file = list(uploaded.keys())[0]  # uploaded Excel filename
cv_file = list(cv_upload.keys())[0]    # uploaded CV filename

# === STEP 2: Load Excel ===
df = pd.read_excel(excel_file)
df.columns = df.columns.str.strip()  # remove spaces in headers

print("âœ… Data Loaded Successfully!\n")
print(df.head())

# === STEP 3: Gmail Settings ===
sender_email = "wattoo7460@gmail.com"
app_password = "######################"   # Gmail App Password (keep secret âš ï¸)

# === STEP 4: Send Function ===
def send_email(to_email, company_name):
    try:
        msg = MIMEMultipart()
        msg['From'] = sender_email
        msg['To'] = to_email
        msg['Subject'] = f"Job Application - Data Scientist Role at {company_name}"

        body = f"""\
        Dear HR at {company_name},

        I hope this email finds you well.
        My name is Ali Nawaz, and I am a Machine Learning Engineer.
        I have been learning and working in this field for over a year, and I am now seeking internship opportunities to gain hands-on industry experience.

        I would be grateful if you could kindly consider me for any available internship positions.
        Please find my CV attached for your review.

        Best regards,
        Ali Nawaz
        """
        msg.attach(MIMEText(body, 'plain'))

        # Attach CV
        with open(cv_file, "rb") as attachment:
            part = MIMEBase('application', 'octet-stream')
            part.set_payload(attachment.read())
            encoders.encode_base64(part)
            part.add_header('Content-Disposition', f'attachment; filename={cv_file}')
            msg.attach(part)

        # Send Mail
        server = smtplib.SMTP('smtp.gmail.com', 587)
        server.starttls()
        server.login(sender_email, app_password)
        server.send_message(msg)
        server.quit()

        print(f"âœ… Email sent to {to_email} ({company_name})")

    except Exception as e:
        print(f"âŒ Failed to send email to {to_email}. Error: {e}")

# === STEP 5: Loop Through Excel ===
for index, row in df.iterrows():
    company = row['Company_Name']
    hr_email = row['HR_Contact']

    if pd.notna(hr_email) and "@" in str(hr_email):  # only valid emails
        send_email(hr_email, company)
```
